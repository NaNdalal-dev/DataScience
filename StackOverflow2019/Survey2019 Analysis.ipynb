```python
import pandas as pd
```

```python
survey_df=pd.read_csv('surveyInsights/survey_results_public.csv',index_col='Respondent')
pd.set_option('display.max_columns',len(survey_df.columns))
survey_df.head()
```

```python
schema_df=pd.read_csv('surveyInsights/survey_results_schema.csv',index_col='Column')
pd.set_option('display.max_rows',85)
schema_df
```

```python
survey_df.head(3)
```

```python
schema_df.loc['Hobbyist','QuestionText']
```

```python
from collections import Counter
from matplotlib import pyplot as plt
hobbyCounter=Counter()
for answers in list(survey_df['Hobbyist']):
    hobbyCounter.update([answers])
#print(hobbyCounter)
yes,no=hobbyCounter['Yes'],hobbyCounter['No']
print('Number of people who code as hobby:',yes)
print('Number of people who don\'t code as hobby:',no)
print('Total',yes+no)
plt.style.use('fivethirtyeight')
plt.pie([yes,no],labels=['yes','No'],shadow=True,wedgeprops={'edgecolor':'black'},autopct='%1.1f%%')
plt.title('Coding as Hobby')
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
from matplotlib import pyplot as plt
country_counter=Counter()
for country in survey_df['Country']:
    country_counter.update([str(country)])
country_name=[]
participants=[]
for country,Totalpartcipants in country_counter.most_common(len(country_counter)):
    country_name.append(country)
    participants.append(Totalpartcipants)
SurveyCountry={
    'Country':country_name,
    'Participants':participants
}
SurveyCountry=pd.DataFrame(SurveyCountry)
SurveyCountry.to_csv('StackOverFlowCountrySurvey.csv')
```

```python
SurveyCountry.head(5)
```

```python
country=list(SurveyCountry['Country'].head(15))
country.reverse()
participants=list(SurveyCountry['Participants'].head(15))
participants.reverse()
plt.style.use('fivethirtyeight')
plt.title('Top 15 countries which participated in StackOverflow survey 2019')
plt.xlabel('Participants')
plt.barh(country,participants)
plt.show()
```

```python
from collections import Counter
jobsat=Counter()
for i in ((survey_df['JobSat'])):
    jobsat.update([str(i)])
jobsat
```

```python
items=list(jobsat)
```

```python
values=[]
for i in range(len(items)):
    values.append(jobsat[items[i]])
values
```

```python
items.reverse()
values.reverse(
)
JobSatisfaction={
    'JobSatisfaction':items,
    'Response':values
}
jobsat_df=pd.DataFrame(JobSatisfaction)
jobsat_df.to_csv('JobSatisfactionData.csv')
jobsat_df
```

```python
plt.title('Survey on job Satisfaction')
plt.style.use('fivethirtyeight')
plt.pie(jobsat_df['Response'],labels=jobsat_df['JobSatisfaction'],colors=['red','green','yellow','grey','lightblue','lightpink'],wedgeprops={'edgecolor':'black'},autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
from collections import Counter
os=Counter()
for sys in survey_df['OpSys']:
    os.update([str(sys)])
osname=list(os)
users=[]
for i in osname:
    users.append(os[i])
```

```python
osname[2],osname[3]=osname[3],osname[2]
users[2],users[3]=users[3],users[2]
```

```python
OSSurvey={
    'OS Name':osname,
    'Number of users':users
}
os_df=pd.DataFrame(OSSurvey)
os_df.to_csv('OS_Survey.csv')
os_df
```
