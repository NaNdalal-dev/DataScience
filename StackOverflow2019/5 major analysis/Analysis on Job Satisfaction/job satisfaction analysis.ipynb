```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Analysis on job satisfaction

```python
jobsat_df=pd.read_csv('JobSatisfactionData.csv')
jobsat_df
```

```python
plt.style.use('fivethirtyeight')
plt.title('Survey on job Satisfaction')
plt.style.use('fivethirtyeight')
plt.pie(jobsat_df['Response'],labels=jobsat_df['JobSatisfaction'],colors=['red','green','yellow','grey','lightblue','lightpink'],wedgeprops={'edgecolor':'black'},autopct='%1.1f%%',shadow=True)
plt.show()
```
