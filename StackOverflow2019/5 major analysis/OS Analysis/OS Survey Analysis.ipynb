# Analysis on Operating Systems

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
os_df=pd.read_csv('OS_Survey.csv')
os_df
```

```python
plt.style.use('fivethirtyeight')
plt.title('Most used Operating systems')
plt.pie(os_df['Number of users'],labels=os_df['OS Name'],colors=['yellow','purple','grey','blue','lightpink'],wedgeprops={'edgecolor':'black'},autopct='%1.1f%%',shadow=True)
plt.show()
```
