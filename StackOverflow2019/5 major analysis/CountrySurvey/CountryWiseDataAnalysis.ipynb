```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
SurveyCountry=pd.read_csv('StackOverFlowCountrySurvey.csv')
SurveyCountry.head(10)
```

```python
country=list(SurveyCountry['Country'].head(10))
country.reverse()
participants=list(SurveyCountry['Participants'].head(10))
participants.reverse()
plt.style.use('fivethirtyeight')
plt.title('Top 10 countries which participated in StackOverflow survey 2019')
plt.xlabel('Participants')
plt.barh(country,participants)
plt.show()
```
