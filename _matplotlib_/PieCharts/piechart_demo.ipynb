```python
from matplotlib import pyplot as plt
slices=[50,25,10,5,5]
#just demo of pie diagram
plt.pie(slices)
plt.title('Colors chart')
plt.tight_layout()
```

```python
#lets labels to the pie chart
labels=['color1','color2','color3','color4','color5']
plt.title('Colors chart')
plt.pie(slices,labels=labels)
plt.tight_layout()
```

```python
#lets add color in borders 
#we use wedgeprops to add color in the border
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
plt.title('Color chart')
plt.tight_layout()
```

```python
#lets remove a pie from chart say color3
#we use explode parameter to remove any pie from pie chart
#Also add shadow to the parameter
explodes=[0,0,0.1,0,0]
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},explode=explodes,shadow=True)
plt.title('Color chart')
plt.tight_layout()
```

```python
#To know the percentage of each color we pass autopct parameter
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},explode=explodes,shadow=True,autopct='%1.1f%%',startangle=180)
plt.title('Color chart')
plt.tight_layout()
```
