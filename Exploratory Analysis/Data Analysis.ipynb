## Importing necessary modules

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

## Readind iris data 

```python
iris_df=pd.read_csv('iris.csv')
iris_df.head()
```

## classifying species

```python
iris_setosa=iris_df[iris_df['species']=='Iris-setosa']
iris_versicolor=iris_df[iris_df['species']=='Iris-versicolor']
iris_virginica=iris_df[iris_df['species']=='Iris-virginica']
```

### Univariate Analysis

```python
plt.style.use('fivethirtyeight')
for i in range(len(iris_df.columns)-1):
    plt.plot(iris_setosa[iris_setosa.columns[i]],np.zeros_like(iris_setosa[iris_setosa.columns[i]]),'o',label='setosa',color='red')
    plt.plot(iris_versicolor[iris_versicolor.columns[i]],np.zeros_like(iris_versicolor[iris_versicolor.columns[i]]),'o',label='versicolor',color='green')
    plt.plot(iris_virginica[iris_virginica.columns[i]],np.zeros_like(iris_virginica[iris_virginica.columns[i]]),'o',label='virginica',color='blue')
    plt.title(iris_df.columns[i])
    plt.legend()
    plt.show()
```

## bivariate analysis

```python
for i in range(len(iris_df.columns)-2):
    print([iris_df.columns[i]],[iris_df.columns[i+1]])
```

### bivariate analysis using seaborn

```python
for i in range(len(iris_df.columns)-2):
    sns.FacetGrid(iris_df,hue='species',size=5).map(plt.scatter,iris_df.columns[i],iris_df.columns[i+1]).add_legend()
    plt.show()
```

### multivariate analysis

```python
sns.pairplot(iris_df,hue='species',size=5)
plt.show()
```
