```python
import pandas as pd
from matplotlib import pyplot as plt
import ST_line as st #my own module
```

```python
#recent covid cases in india
df=pd.read_csv('covid_India_withdates.csv')
df.tail()
```

```python
df1=pd.read_csv('covid_India.csv')
```

```python
plt.style.use('fivethirtyeight')
plt.title('Covid confirmed cases in india')
plt.xlabel('Days(from jan 22-march 30)')
plt.ylabel('confirmed cases')
plt.plot(df1.day,df1.confirmed,label='confirmed cases')
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.title('Covid death cases in india')
plt.xlabel('Days(from jan 22-march 30)')
plt.ylabel('deaths')
plt.plot(df1.day,df1.deaths,label='deaths',color='red')
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.title('Covid recovered cases in india')
plt.xlabel('Days(from jan 22-march 30)')
plt.ylabel('recovered cases')
plt.plot(df1.day,df1.recovered,label='recovered',color='y')
plt.legend()
plt.show()
```

```python
covid_ind=pd.read_csv('Corona_day_to_day.csv')
covid_ind.tail()
```

```python
new_con=st.obs(covid_ind.day,covid_ind.confirmed,predict=[70,71,72])
new_deaths=st.obs(covid_ind.day,covid_ind.deaths,predict=[70,71,72])
new_recovered=st.obs(covid_ind.day,covid_ind.recovered,predict=[70,71,72])
```

```python
t=0
print('Predicted cases for next 3 days(in india):')
for i in range(3):
    print(round(new_con[i]))
    t+=round(new_con[i])
print('Total',t)
t=0
print('\n\nPredicted death cases for next 3 days(in india):')
for i in range(3):
    print(round(new_deaths[i]))
    t+=round(new_deaths[i])
print('Total',t)
t=0
print('\n\npredicted recovered cases for next 3 days(in india):')
for i in range(3):
    print(round(new_recovered[i]))
    t+=round(new_recovered[i])
print('Total',t)
```
