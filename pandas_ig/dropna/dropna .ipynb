```python
from pandas import read_csv
```

```python
df = read_csv('weather_data.csv')
df
```

### Dropping "unknown" column 

```python
df.drop(labels=['unknown'],inplace=True,axis=1)
#Note
#"inplace=True" will permanently remove the column
#axis = 1 indicates column where as axis = 0 is row
```

```python
df
```

### Dropping the rows that contains all NaN values

```python
df.dropna(how='all',axis=0)
```

### Dropping the rows that contains  NaN values

```python
df.dropna(how="any",axis=0) #This is the default parameter
```

### Dropping the rows that contains 2 or more NaN values 

```python
df.dropna(thresh=2,axis=0)
```
