```python
import pandas as pd
df=pd.read_csv('weather_data.csv',parse_dates=['day'])
df
```

```python
#we can observe that the above has some missing values(i.e. NaN)
```

```python
df.set_index('day',inplace=True)
```

```python
 df
```

```python
#lets ignore the rows haing NaN by using dropna function
print(df)
drop=df.dropna()
drop
```

```python
#lets ignore a row having no data using no data
print(df)
dropall=df.dropna(how='all')
dropall
```

```python
#lets have  treshhold values
print(df)
thresh_by1=df.dropna(thresh=1)
thresh_by2=df.dropna(thresh=2)
print(thresh_by1)
print(thresh_by2)
```

```python
#lets fill the values of NaN with 0
print(df)
fill_zero=df.fillna(0)
fill_zero
```

```python
print(df)
#Lets make the table more appropiate using dictionary
fill_appropiate=df.fillna({
    
    'temperature':0,
    'windspeed':0,
    'event':'No Event'
    
    
})
fill_appropiate
# :-)
```

```python
print(df)
#lets carry forward the values
forward_df=df.fillna(method='ffill')
forward_df
```

```python
print(df)
#lets backward the values
backward_fill=df.fillna(method='bfill')
backward_fill
```

```python
print(df)
pad_fill=df.fillna(method='pad')
pad_fill
```

```python
print(df)
#horizontal fill backward using axis
axis_fill=df.fillna(method='ffill',axis='columns')
axis_fill
```

```python
print(df)
#horizontal fill forward using axis
axis_fill=df.fillna(method='ffill',axis='columns')
axis_fill
```

```python
print(df)

back=df.fillna(method='backfill')
back
```

```python
limit1=df.fillna(method='ffill',limit=1)#forwards the value 1 time(limit)
print(limit1)
limit2=df.fillna(method='ffill',limit=2)#forwards the value twice
print(limit2)
```
