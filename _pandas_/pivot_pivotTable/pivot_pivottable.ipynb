```python
#pivot and pivot_tables
import pandas as pd
df1=pd.read_csv('weather.csv',date_parser='date')
df1
```

```python
#Pivot():It helps to reshape or transform column to row 
#lets look at the following example
df1.pivot(index='date',columns='city')
```

```python
#in the above pivot table if you want to humidity as table then
df1.pivot(index='date',columns='city',values='humidity')
```

```python
#Pivot_table():It is used to summarise and aggregate(mean,sum,diff,prod...etc) the table
df2=pd.read_csv('weather2.csv',date_parser='date')
df2
```

```python
#converting date column to date
df2.date=pd.to_datetime(df2.date)
type(df2.date[0])
```

```python
#lets summarise the table with average humidity and temperatute on both dates
df2.pivot_table(index='date',columns='city',aggfunc='mean')
```
