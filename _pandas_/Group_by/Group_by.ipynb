```python
import pandas as pd
df=pd.read_csv('weather_by_cities.csv')
df
```

```python
#lets separate the by cities 
g=df.groupby('city')
for city,city_df in g:
    print(city)
    print(city_df)
```

```python
#lets get the data of newyork
nyc_df=g.get_group('new york')
nyc_df
```

```python
#highest temperature and wind speed
gmax=g.max()
gmax
```

```python
#lowest temperature and windspeed
gmin=g.min()
gmin
```

```python
#Average temperature and wind speeed
gmean=g.mean()
gmean
```

```python
#statistics
gdesc=g.describe()
gdesc
```

```python
import matplotlib.pyplot as plt
g.plot()
```
