```python
import pandas as pd
df=pd.read_excel('allpubgwpnstats.xlsx')
df
```

```python
#lets group the weapons 
df.dropna()
wpn_type=df.groupby('WeaponType')
for types,weapons in wpn_type:
    print('Weapon category :',types)
    print()
    print('Statistics :\n',weapons)
    print('\n\n')
```

```python
#let us the type of ammo that weapons use
ammo=df.groupby('Ammo')
for ammotype,guns in ammo:
    print('Ammo Type : ',ammotype)
    print()
    print('Statistics :\n',guns)
    print('\n\n')
```

```python
#lets us know the weapon with highest damage


dmax=   df[['Weapon','Damage']][df.Damage==df.Damage.max()]
dmax
```

```python
#lets ud know the weapon with highest speed
import numpy as np
df.fillna({
    'Time to kill(in sec)':np.NaN
})
print('Weapon with highest damage:')

df[['Weapon','Time to kill(in sec)']][df['Time to kill(in sec)']==df['Time to kill(in sec)'].min()]
```
