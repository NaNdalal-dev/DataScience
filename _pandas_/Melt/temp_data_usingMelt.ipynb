```python
#Melt: This method in pandas is used to  transform or reshape data
#lets demonstrate
import pandas as pd
df=pd.read_csv("weather.csv")
df
```

```python
df1=pd.melt(df,id_vars='day')
df1
```

```python
#since melt() is putting default column names such as variable and value  lets change to our wish
df1=pd.melt(df,id_vars='day',var_name='temperature',value_name='city')
df1
```

```python
#lets us take a individual data
df1=pd.melt(df,id_vars='day',var_name='city',value_name='temperature')

df1[df1['city']=='chennai']
```
