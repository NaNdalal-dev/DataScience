```python
import pandas as pd
df=pd.read_csv('aapl_no_dates.csv')
df
```

```python
#we can clearly see that above data has no dates which makes no sense
#so to put dates as index  we use date_range() which consists of start and end day(or periods)
dti=pd.date_range(start='2017-01-01', end='2017-01-31',freq='b')
#'b' indicates business days which excludes weekends(sat & sunday)
dti
```

```python
#now lets pass the above dti(DatetimeIndex) in df 
df.set_index(dti,inplace=True)
df
```

```python
#now the tabular data consists dates 
```
